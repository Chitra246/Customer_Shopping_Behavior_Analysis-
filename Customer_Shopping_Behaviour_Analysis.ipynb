```python
#Finding the median of the each catagory to fill the Null values. 
#By doing this for each catagory we are removing the chances of data biasness.
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Changingt the Columns Name Formate to Snake Casing for easy accebility and uniformity
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# Creating age_group column with reference to the cutomer age given
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels) #qcut divides a columns into subcatagories by no. of subcategory 'q'
```

```python
# Create columns purchase_frequency_days from purchase_frequesncy which is in text format to integer
#to make calculations easy 
frequency_mapping={'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Quaterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90,
} # Creating Dic. to map the numbers.
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
